#### Import the necessary libraries here:

```python
# your code here
import requests
import json
import pandas as pd
from pandas import json_normalize
```

```python
response = requests.get('https://api.github.com/events', 
                        headers={'Accept': 'application/vnd.github.v3+json'})
response.request.url


data = json_normalize(response.json())
```

```python
f = open ('token_demo_ironhack.txt', 'r')
github_token = f.read()
f.close()
```

```python
response = requests.get('https://api.github.com/user', headers={'Accept': 'application/vnd.github.v3+json',
                                                               'Authorization': f'token {github_token}'})
results = response.json()
results
```

```python
response = requests.post('https://api.github.com/user/emails',
                        headers={'Accept': 'application/vnd.github.v3+json',
                                'Authorization' : f'token {github_token}'},
                        data= json.dumps({'emails': ['pholasajustine@yahoo.fr']}))
```

```python
response.json()
```

```python
response = requests.delete('https://api.github.com/user/emails',
                        headers={'Accept': 'application/vnd.github.v3+json',
                                'Authorization' : f'token {github_token}'},
                        data= json.dumps({'emails': ['pholasajustine@yahoo.fr']}))
```

```python
response.content
```

#### Question 1: Find the questions and answers of last month.

```python
from stackapi import StackAPI
SITE = StackAPI('stackoverflow')
```

```python
# your code here
questions = SITE.fetch('questions', fromdate=1612137600, todate=1614470400)
```

```python
questions['items']
```

```python
answers = SITE.fetch('answers', fromdate=1612137600, todate=1614470400)
```

```python
answers['items']
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
most_voted_question_today = SITE.fetch('questions', 
                                       fromdate=1616803200, todate=1616803200,min=5, 
                                       tagged='python', sort='votes')
```

```python
most_voted_question_today['items']
```
